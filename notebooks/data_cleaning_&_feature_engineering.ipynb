ADD FEATURE ENGINEERING

Delivery delay in days (Actual − Estimated)

```python
orders["delivery_delay_days"] = (orders["order_delivered_customer_date"] - orders["order_estimated_delivery_date"]).dt.days
```

On-time delivery indicator (1 = On Time, 0 = Delayed)

```python
orders["on_time_flag"] = ( orders["delivery_delay_days"] <= 0 ).astype(int)
```

Days from purchase to carrier pickup

```python
orders["processing_days"] = ( orders["order_delivered_carrier_date"] 
                - orders["order_purchase_timestamp"]).dt.days
orders["processing_days"] = orders["processing_days"].clip(lower=0)
```

Days from carrier pickup to customer delivery

```python
orders["shipping_days"] = ( orders["order_delivered_customer_date"] - orders["order_delivered_carrier_date"] ).dt.days.clip(lower=0)
```

Number of items per order

```python
basket_size = (order_items.groupby("order_id").size().reset_index(name = "basket_size"))

orders = orders.merge(basket_size , on = "order_id", how = "left")
```

Purchase month

```python
orders["purchase_month"] = orders["order_purchase_timestamp"].dt.month
```

Purchase weekday

```python
orders["purchase_weekday"] = orders["order_purchase_timestamp"].dt.day_name()
```

CREATE ANALYSIS TABLE

```python
analysis = orders.copy()
```

```python
analysis = analysis.merge( customers[["customer_id", "customer_state"]],
    on="customer_id",
    how="left" )

analysis = analysis.merge(order_reviews[["order_id", "review_score"]],
                    on = "order_id", how = "left")

analysis = analysis.merge( order_items[ ["order_id", "product_id", "seller_id", "price", "freight_value"]], on="order_id", how="left")

analysis = analysis.merge( product[ ["product_id", "product_category_name"]],
    on="product_id", how="left" )

analysis = analysis.merge( sellers[ ["seller_id", "seller_state"] ],
    on="seller_id", how="left" )
```

```python
analysis = analysis.merge(translation_name[["product_category_name", "product_category_name_english"]], on = "product_category_name", how = "left")
```
